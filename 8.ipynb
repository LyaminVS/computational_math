```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# метод Рунге-Кутта 1 порядка
def first_order(f, u_0, h, interation_number):
    u = np.zeros((interation_number + 1, len(u_0)))
    u[0] = u_0
    for i in range(interation_number):
        u[i + 1] = u[i] + h * f(u[i])
    return u
    
# метод Рунге-Кутта 2 порядка    
def second_order(f, u_0, h, interation_number):
    u = np.zeros((interation_number + 1, len(u_0)))
    u[0] = u_0
    for i in range(interation_number):
        u[i + 1] = u[i] + h / 2 * f(u[i]) + h / 2 * f(u[i] + h * f(u[i]))
    return u
    
# метод Рунге-Кутта 3 порядка
def third_order(f, u_0, h, interation_number):
    u = np.zeros((interation_number + 1, len(u_0)))
    u[0] = u_0
    for i in range(interation_number):
        k_1 = f(u[i]) * h
        k_2 = f(u[i] + k_1 / 2) * h
        k_3 = f(u[i] - k_1 + 2 * k_2) * h
        u[i + 1] = u[i] + (k_1 + 4 * k_2 + k_3) / 6
    return u

# метод Рунге-Кутта 4 порядка
def fourth_order(f, u_0, h, interation_number):
    u = np.zeros((interation_number + 1, len(u_0)))
    u[0] = u_0
    for i in range(interation_number):
        k_1 = f(u[i]) * h
        k_2 = f(u[i] + k_1 / 2) * h
        k_3 = f(u[i] + k_2 / 2) * h
        k_4 = f(u[i] + k_3) * h
        u[i + 1] = u[i] + (k_1 + 2 * k_2 + 2 * k_3 + k_4) / 6
    return u
    
    
def f(u):
    v = u[0]
    w = u[1]
    
    return np.array([v + w, v ** 2 - w ** 2])
```

```python
# функция для построения графиков в зависимости от шага сетки
def plots(h):
    
    u_0 = np.array([1, 2])
    iteration_number = int(1 / h)
    
    u_1 = first_order(f, u_0, h, iteration_number)
    u_2 = second_order(f, u_0, h, iteration_number)
    u_3 = third_order(f, u_0, h, iteration_number)
    u_4 = fourth_order(f, u_0, h, iteration_number)
    
    plt.plot(np.arange(0, 1 + h, h), u_1[:, 0], label="1 порядок")
    plt.plot(np.arange(0, 1 + h, h), u_2[:, 0], label="2 порядок") 
    plt.plot(np.arange(0, 1 + h, h), u_3[:, 0], label="3 порядок") 
    plt.plot(np.arange(0, 1 + h, h), u_4[:, 0], label="4 порядок") 
    plt.title("v(t)")
    plt.xlabel("t")
    plt.ylabel("v")
    plt.legend()
    plt.show()
    
    plt.plot(np.arange(0, 1 + h, h), u_1[:, 1], label="1 порядок") 
    plt.plot(np.arange(0, 1 + h, h), u_2[:, 1], label="2 порядок")  
    plt.plot(np.arange(0, 1 + h, h), u_3[:, 1], label="3 порядок")
    plt.plot(np.arange(0, 1 + h, h), u_4[:, 1], label="4 порядок") 
    plt.title("w(t)")
    plt.xlabel("t")
    plt.ylabel("w")
    plt.legend()
    plt.show()
```

```python
# графики для h = 0.01
plots(0.01)
```

```python
# графики для h = 0.1
plots(0.1)
```

```python
# графики для h = 0.0001
plots(0.0001)
```
